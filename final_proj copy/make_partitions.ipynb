```python
def listabs(directory):
    '''Returns the absolute path of all items in a directory
    
    Parameters
    ----------
    directory : path to directory'''
    return [os.path.join(directory, filename) for filename in os.listdir(directory)]
```

```python
phyles = listabs('.')
```

```python
splits = [re.split('(\d+)', phy) for phy in phyles]
fill = len(str(splits[-1][1]))

for split in splits:
    split[1] = split[1].zfill(3)
    
newphyles = [''.join(split) for split in splits]
```

```python
for (old, new) in zip(phyles, newphyles):
    shutil.move(old, new)
```

```python
parts = []
i=1
length = 0
for phy in newphyles:
    with open(phy, 'r') as p:
        locuslen = int(p.readline().split()[1])
        length+=locuslen
        parts.append([i, length])
        i+=locuslen   
```

```python
parts
```

```python
partlines = []
for (locus, part) in zip(newphyles, parts):
    locusind = re.split('[/.]', locus)[2]
    partlines.append(('DNA, %s=%d-%d\n') % (locusind, part[0], part[1]))
```

```python
with open('cbbproj.phy.partions', 'w+') as p:
    p.writelines(partlines)
```
